## Nuestro primer modelo de Machien Learning

Un modelo de Machine Learning, a veces, tambien es llamado modelo de prediccion. Cuando escuchamos un modelo de prediccion muchas veces creemos que estamos hablando de algo que solo hace una afirmacion segun lo que se cree, pero no es asi. Un modelo de prediccion funciona a partir de tomar los datos recopilados y aplicarles una ecuacion matematica tan simple como la ecuacion de la pendiente (y = mx + b). Si seguimos lo resultado que nos da esta ecuacion, entonce si le damos los datos que no tenemos pero podemos preveer, nos dara una respuesta que podemos utilizar como prediccion.

Para comenzar, primero necesitamos saber que funciones, columnas, ocuparemos.

```python
# funciones
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

Ya que tenemos las funciones, ahora las ponemos a prueba con el modelo. Ocuparemos un modelo de Arbol de Deciciones Regresivo.

```python
from sklearn.tree import DecisionTreeRegressor
```

Dentro del metodo que ocuparemos, necesitaremos dos paremetros, los predictores y el objetivo. Los predictores, seran la variables X, seran aquellos en el que nos basaremos para hacer la prediccion. Y la prediccion sera y, el precio.

```python
y = df.Price
X = df[melbourne_features]
```

Ahora, hacer la formula que ocuparemos para hacer la prediccion ya esta hecha por el metodo que ocuparemos, asi que, solo necesitamos "ajustar" el modelo.

```python
melbourne_model = DecisionTreeRegressor()
```

```python
melbourne_model.fit(X, y)
```

Ya esta ajustado. Ahora queda saber que prediccion nos puede dar.

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

Cunado tenemos las columnas que utilizaremos para hacer la prediccion, los nombramos como "Labels" y a la prediccion o la columna que nos ayudara a entrenar la prediccion sera el objetivo o "Target". Lo que hemos hecho fue entrenar el modelo con los datos que ya teniamos para poder hacer una prediccion, lo que hicimos fue darle de nuevo datos con lo que a habia entrenado pero ahora hicimos que volviera a hacer una prediccion de las primeras 5 entradas, en teoria no deberia de haber error en ello pues se supone que ya sabe el modelo que resultados obtendriamos si tomamos en cuenta su entrenamiento, pero no es asi del todo, ahora.

El modelo de prediccion hace una prediccion basado en un entrenamiento previo que tuvo con datos que le habiamos dado y no de datos alazar, es decir, si le damos datos distintos podemos tener otras prediciones. Aunque hagamos funcionar un modelo de prediccion esto no quiere decir que realmente funciona, seria todo un milagro que la prediccion que queremos sea exacta, es decir, no paresida o similar al resultado real, sino completamente iguales. Cuando se entrena un modelo suele darse dos conjuntos de objetivos, el primer conjunto de objetivos es para entrenarla y el segundo para poenerlo a prueba, cuando se pone a pruba se evalua con una funcion, llamada Error Absoluto Medio (Median Absolute Error o MAE) que lo que hace es comparar los resultados que nos da una prediccion con un modelo entrenado con los resultados reales. Basicamente, si el error es 0 entonces esto nos indicaria que el error es exacto, pero, en la mayoria de ocaciones, aunque un modelo sea lo mas preciso posible, el error suele ser disitnto a cero, es decir, puede ser de hasta un error de 2 hasta 10, que entre mas pequeño sea quiere decir que mas exacto es. Nuestro trabajo por el momento sera crear un modelo que nos ayude a encontrar un MAE cercano a 10.

## Validacion del Modelo

Como se mensiono antes, la validacion del modelo nos dira que tan exacto es nuestro modelo. Esto no quiere decir que la funcion que utilizemos funcione mal, sino que hemos elegido los atributos correctos y que sirven para hacer una prediccion, pero, si nuestro dataset es de 10 atributos y solo usamos la mitad, es obvio que el modelo no sera tan exacto, para eso hay que hacer algo con los detalles incorrectos antes de usarlos.

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

La media de error es demaciado grande, es un hecho que para nada nuestro modelo es exacto, todo lo contrario. Aunque por ahora, no es necesario corregir esto, solo lo ocupamos para poder entender como funcionan los metodos que usamos de la libreria sklearn. Para poder hacer un mejor trabajo, podriamos hacer trabajar mejor nuestro modo de usar el dataset. Podriamos dividir nuestro dataset en un conjunto de entrenamiento y uno de validacion.

### Conjunto de entrenamiento y validacion

```python
from sklearn.model_selection import train_test_split
# Dividimos los conjuntos X e y para hacer la divicion
X_train, X_test, y_train, y_test = train_test_split(X, y)
# volvemos a hacer el modelo ahora con los conjuntos de validacion
melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(X_train, y_train)


# hacemos la prediccion ahora con el conjunto de validacion
prediction = melbourne_model.predict(X_test)
# y hacemos la validacion
print(mean_absolute_error(prediction, y_test))
```

Ahora vemo que le modelo de prediccion es mas grande, pero eso se debe a que usamos la mitad de los datos para hacer el conjunto de datos.
