# Scaling and Noramalitation (escalando y normalizando)
Ya hemos visto que tenemos datos continuos y discretos, cualitativos y cuantitativos. Ahora, vemos que con aquellos valores que creemos ya estan bien tal como estan, ahora veremos que pueden estar mejor.

En muchos datasets tenderemos columnas cuyos valores ronden intervalos muy grandes y muy chicos, algunos muy separados entre si y otros apenas diferentes. Eso puede crear problemas de ejecucion, ya que de esa forma python hace calculos muy disjuntos y puede tambien dificultar como procesamos la informacion. Para arreglar todos esos problemas existe la escalabilidad. **Escalar** los datos es una practica muy buena en el analisis de datos, ya que basicamente lo que que hacemos es hacer que todos los datos se lean en un rango de numeros, parecidos pero no iguales. 

Imaginemos que tenemos valores de 100 registros, y esos 100 registros tienen 2 columnas, edad y cuanto dinero ganan. La edad se lee desde los 13 años hasta los 80, mientra que el dinero se cuenta desde los $0 hasta los millones. Intuitivamente observamos que los valores no se equiparam pues haciendo escalacion de datos hara que esa diferencia de rango no sea tan distinta, es decir, se leeran de 1 a 10 por ejemplos en ambas columnas.

```python
# ejemplo
import pandas as pd
import numpy as np

from scipy import stats

from mlxtend.preprocessing import minmax_scaling

import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(10)
```

```python
original_data = np.random.exponential(size = 1000)

scaled_data = minmax_scaling(original_data, columns = [0])

fig, ax = plt.subplots(1, 2, figsize = (15, 3))
sns.histplot(original_data, ax = ax[0], kde = True, legend = False)
ax[0].set_title("original_data")
sns.histplot(scaled_data, ax = ax[1], kde = True, legend = False)
ax[1].set_title("Scaled_data")
plt.show()
```

Vemos que el rango que va de 0 hasta casi 12 ahroa va del 0 al 1.

## Normalizacion

Escalando solo cambia el rango de tu data, la normalizacion es mas radical, hace que tu informacion se distribuya de forma "*Nomral*" (de no saber que significa, entonces buscar *distribucion normal*).

```python
normalized_data = stats.boxcox(original_data)

fig, ax = plt.subplots(1, 2, figsize = (15, 3))
sns.histplot(original_data, ax = ax[0], kde = True, legend = False)
ax[0].set_title("Original data")
sns.histplot(normalized_data, ax = ax[1], kde = True, legend = False)
ax[1].set_title("Normalized data")
```
