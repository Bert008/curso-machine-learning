## Sobreajuste y desajuste

Cuando hacemos un modelo de entrenamiento este debe de aprender para poder trabajar, pero, en el proceso de entrenamiento este puede entrenar mal. Cuando hacemos un modelo que trata de aprender este lo hara de la informacion que le demos, asi que si le damos informacion erronea este lo entendera mal y cuando haga una prediccion entonces no se acercara a un resultado real. Tambien lo hace al contrario, puede que los resultados que aparescan sean parecidos a los correctos, pero entramos en un concepto que se concidera mas peligros y es el sobreajuste. El sobreajuste "generaliza" segun lo que aprenda y sus resultados pueden ser mal ententidos, habria que analizarlos bien ya que es dificil identificar de un resultado valido a un resultado generalizado. Un modelo que se sobreajusta por lo general tiende a dar resultados bastante predecibles y de un rango cercano entre resultados. Es decir, la probabilidad de que nos pueda dar un resultado bastante sorprendente o poco comun es poca.

Aunque ya he dicho que el soberajuste se debe a la entrada de datos, tambien se debe a como el modelo los aprender, el modelo tambien tiene gran parte de ese ajuste. Por ejemplo, el modelo que tomamos es DecisionTreeRegressor o Arbol de deciciones regresivo. Este modelo toma los datos y los divide en ojas, entre mas ojas tenga mas deciciones hace y entre mas decisiones haga mejor criterio tendra sobre que hacer con cada dato. Esto es nesesario de etender ya que cuando usamos la funcion este nos da mas parametros para analizar mejor la informacion.


```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, X_train, X_valid, y_trian, y_valid):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(X_train, y_train)
    preds_val = model.predict(X_valid)
    mae = mean_absolute_error(y_valid, preds_val)
    return (mae)
```

```python
import pandas as pd
melbourne_data = pd.read_csv("data/melb_data.csv")
# eliminamos la informacion nula (null)
filtered_melbourne_data = melbourne_data.dropna(axis = 0)

y = filtered_melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']
X = filtered_melbourne_data[melbourne_features]

from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 0)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, X_train, X_valid, y_train, y_valid)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
```

Vemos que las 500 ojas son las mas optimas para este ejemplo, ya que si hacemos mas de 500 esntonces el error se hace mas grande.
