# Missing Values

Los valores perdidos, o missing values, son aquellos de los cuales no existe una entrada, es decir, por alguna razon no estan. Hay que diferenciar entre valores perdidos o nulos y valores reales, muchas veces los principiantes suelen confundir un valor 0 con un valor valor nulo, pero no, un valor 0 es un valor, es decir, fue puesto ahi al proposito, en cambio un valor nulo no fue puesto ahi al proposito. Dependiendo de que tipo de datos se esten recabando un valor nulo tambien pordria decirnos algo de informacion, pero esto esta muy ligado a lo que se este pidiendo. Por el momento, un valor nulo, que en la computadora se observa como Null, tiene que tratarse con cuidado dependiendo de lo que buscamos.

Existen varias formas de tratar un un valor nulo:
* Tratamos con el valor (lo remplazamos o imputacion).
* Lo eliminamos.

```python
import pandas as pd
melbourne_data = pd.read_csv("data/melb_data.csv")
y = melbourne_data.Price

melb_predictors = melbourne_data.drop(['Price'], axis = 1)
X = melb_predictors.select_dtypes(exclude = ['object'])

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# definimos una funcion para poder comparar las diferentes aproximaciones
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators = 10, random_state = 0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

## Eliminamos los datos perdidos (Drop Missig Values) 

```python
# identificamos las columnas con datos perdidos
cols_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()]

# eliminamos las columnas con datos perdidos
red_X_train = X_train.drop(cols_with_missing, axis = 1)
red_X_valid = X_valid.drop(cols_with_missing, axis = 1)

print("MAE de la aprox. 1 (DMV):")
print(score_dataset(red_X_train, red_X_valid, y_train, y_valid))
```

## Imputacion

La imputacion es un metodo mejor que eliminar los datos en algunas ocaciones. Muchas veces es preferible imputar los datos que eliminarlos, ya que si lo eliminamos, podriamos poder gran parte de los registros que ya tenemos. Existen formas de imputar los datos perdidos, la mas usada es remplazar los datos perdidos con el promedio de los datos que si hay.

```python
from sklearn.impute import SimpleImputer

# Imputacion 
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns

print("MAE de aprox. 2 (imputacion")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

Facilmente vemos que la aproximacion de imputacion es mejor que la de solo eliminar los datos que no hay.
