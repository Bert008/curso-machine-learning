# Deep Neural Networks

La muestra pasada, vimos como definiamor una neurona lineal, es decir, solo hay una entrada y una salida. Pero ahora, vermos como definir toda una capa de neuronas.

## Layers

Es comun que las redes neuronales se organicen en capas (layers). Cuando tenemos todo un grupo de capas de unidades lineales en medio de un conjunto de entradas y otro de salidas, entonces obtenemos un capa de dencidad (dense layer). Esta capa es donde se hara todo el procesamiento de la informacion que le daremos a la red, pero, no solo podemos tener una cola capa, podemos tener varias **dense layers**.

## Construyendo un modelo sequencial

Un modelo sequencial, o `Sequential model`, es un modelo que usarmeos para conectar una lista de capas. La primera capa sera el input y la ultima la salida.

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    # las capas oculatas ReLu
    layers.Dense(units=4, activation='relu', input_shape=[2]),
    layers.Dense(units=3, activation='relu'),
    # the linear output layer 
    layers.Dense(units = 1),
])
```
