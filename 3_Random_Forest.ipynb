# Random Forest Regressor

Ya hemos tocado tema con el arbol de decicion, Desicion Tree Regressor, que divide la informacion en "ramas". Estas ramas se dividen mediante preguntas binarias, como si o no, para poder tomar una decision final. Mientras que un bosque (forest) es un conjunto de de arboles que hacen el mismo procedimiento pero el resultado final es el romedio de los arboles en total.

```python
import pandas as pd

melbourne_data = pd.read_csv("data/melb_data.csv")
# eliminamos filas con valores perdidos
melbourne_data = melbourne_data.dropna(axis = 0)
# escojemos el target y labels
y = melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]

from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state = 1)
forest_model.fit(X_train, y_train)
melb_preds = forest_model.predict(X_valid)
print("mean absolute error: ", mean_absolute_error(y_valid, melb_preds))
```

Que si lo comparamos con las pasadas pruebas con el arbol de deciciones, vemos que el resultado es mejor del que esperabamos.
