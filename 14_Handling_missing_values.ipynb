# Handling Missing Values (arreglando valores perdidos)

Ya tocamos el tema de que como manejar los valores categoricos para usarlos en una regresion para hacer una prediccion. Hemos visto que entre mas datos manejemos, es decir tengamos, nuestra prediccion puede ser mas certera.

En algunas ocaciones, existiran valores que no tengan un valor real como tal, su ausencia tambien puede ser interpretada intuitivamente. Los valores perdidos (missing values) tienden a poder ser remplazados o no ser utilizados cuando se trata de analizar los datos. Ya vismo que podemos remplazarlos, aunque tambien existen otros metodos para no quedarnos sin esos datos.

## Entontremos esos valores perdidos

Cuando se trate de dejar valores perdidos, veremos que pandas no permite que las casillas donde el valor deberia de ir no hay nada. Al no haber nada, pandas lo remplazara con NULL el lugar donde deberia de haber algo y cuando tratemos de ver el numero total de valores que hay en una columna entonces todos esos valores Null no formaran parte del resultado.

```python
import pandas as pd
import numpy as np

nfl_data = pd.read_csv("data/NFL Play by Play 2009-2017 (v4).csv")
np.random.seed(0)
```

```python
nfl_data.head()
```

Vemos que hay datos que los marca como NaN (Not a Number). Aquellos datos NaN, a veces NULL, son aquellos que debemos manejar con cuidado. Eso lo hacemos ya que si tenemos un gran porcentaje de datos perdidos, por ejemplo un poco mas del 20%, estariamos perdiendo la misma cantidad de informacion. Ademas, un dato que no existe puede tener muchas expliaciones dependiendo de la informacion que se este pidiendo, ya que en algunas ocaciones es posible que el registro no se haga de forma deliberada y quien halla pedido ese datos no halla hecho algo al respecto cuando se tratara de un caso asi, como dar un valor por defecto en caso de que se elija no dar uno.

## Cuantos valores perdidos tenemos

Ahora, calculemos cuanta informacion no tenemos.

```python
missing_values_count = nfl_data.isnull().sum()

missing_values_count[0:10]
```

Vemos que el valor con mas datos perdidos es `down`.

```python
total_cells = np.prod(nfl_data.shape)
total_missing = missing_values_count.sum()

percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

Vemos que el total de informacion queno hay en el dataset es de casi el 28%, en perspectiva, esa es mucha informacion. Si decidieramos solo no tomar encuenta esos datos que no hay entonces no habria alguna buena prediccion realmente.

## Eliminando la missing data

Supongamos que decidimos que es buena idea eliminar la informacion perdida, entonces, hagamoslo

```python
nfl_data.dropna()
```

Y hemos visto que yan o hay nada. Afortunadamente no hemos eliminado la informacion realmente, solo hemos eliminado la informacion de esa instancia de memoria. La informacion aun existe, pero hemos visto que eliminarla asi solo por que si entonces nos quita toda la informacion, ahora hagamoslo bien.

```python
columns_with_na_dropped = nfl_data.dropna(axis = 1)
columns_with_na_dropped.head()
```

```python
print("Columnas en la informacion original: %d \n" % nfl_data.shape[1])
print("Columns with na's dropped: %d" %columns_with_na_dropped.shape[1])
```

La informacion que quedo es apenas una cuarta parte de la informacion original. Lo que paso con la funcion `.drop` fue que, por defecto, este elimina las filas que contienen valores perdidos, cuyo atributo es `axis = 0`. Con el atributo `axis` cambiado a 1 de la funcion, entonces lo que estmos tomando en cuenta son las columnas, por lo que esta ahora elimina columnas y no renglones, es preferible de esta forma ya que casi todos los registros contienen valores perdidos.

## Rellenando los valores perdidos

Otra opcion facil de manejar es el de darle un valor a aquellos valores que hay. Esto depende de cada conjunto de datos.

```python
subset_nfl_data = nfl_data.loc[:, "EPA": "Season"].head()
subset_nfl_data
```

```python
subset_nfl_data.fillna(0)
```

La funcion `fillna` contiene atributos utiles, para complementar mejor su uso.

```python
subset_nfl_data.fillna(method = 'bfill', axis = 0).fillna(0)
```
